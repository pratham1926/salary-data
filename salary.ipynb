```python
import pandas as pd

data = {'YearsExperience': [1.1, 1.3, 1.5, 2, 2.2, 2.9, 3, 3.2, 3.2, 3.7, 3.9, 4, 4, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6, 6.8, 7.1, 7.9, 8.2, 8.7, 9, 9.5, 9.6, 10.3, 10.5],
        'Salary': [39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189, 63218, 55794, 56957, 57081, 61111, 67938, 66029, 83088, 81363, 93940, 91738, 98273, 101302, 113812, 109431, 105582, 116969, 112635, 122391, 121872]}

df = pd.DataFrame(data)
```

```python
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.show()
```

```python
print(df.isnull().sum())
```

```python
X = df['YearsExperience']
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X.values.reshape(-1, 1), y)
```

```python
predicted_salary = model.predict(X.values.reshape(-1, 1))
```

```python
r_squared = model.score(X.values.reshape(-1, 1), y)
print("R-squared:", r_squared)
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, predicted_salary, color='red', label='Predicted')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
